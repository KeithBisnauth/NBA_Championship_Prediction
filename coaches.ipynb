```python
# Create years list
```

```python
years = list(range(1997, 2024))
```

```python
# Import packages and libraries
```

```python
from bs4 import BeautifulSoup
from io import StringIO
import pandas as pd
```

```python
# Save seasonal coaching data at html files
```

```python
import requests

url = "https://www.basketball-reference.com/leagues/NBA_{}_coaches.html"

for year in years:
    url_1 = url.format(year)
    new_response = requests.get(url_1)

    with (open("playoff/{}.html".format(year), "w+") as f):
        f.write(new_response.text)
```

```python
# Save scraped data as a dataframe
```

```python
df = []
for year in years:
    with open("coaches/{}.html".format(year)) as f:
        page = f.read()
    soup = BeautifulSoup(page, 'html.parser')
    table = soup.find_all(id="NBA_coaches")
    season = pd.read_html(StringIO(str(table)))[0]
    
    df.append(season)
```

```python
df
```

```python
#Concatenate dataframe 
```

```python
coaches = pd.concat(df)
```

```python
coaches
```

```python
# Save as csv
```

```python
coaches.to_csv("coaches.csv")
```
